```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
### 生成标准正态分布的随机数
np.random.seed(12345)
N = 10000
Z = np.random.normal(size=N)
plt.plot(Z[0:-1:100])
plt.show()

mean_Z = np.mean(Z)
std_Z = np.std(Z)
exkur_Z = scipy.stats.kurtosis(Z)
skew_Z = scipy.stats.skew(Z)
print("mean =",mean_Z,"std =",std_Z,"excess kurtosis =", exkur_Z,"skew =",skew_Z)
```

```python
ZH = Z[0:Z.size//2]
mean_ZH = np.mean(ZH)

ZA = np.r_[ZH,-ZH] ## np.r_ 相当于concatinate，也算是一个trick吧
mean_ZA = ZA.mean()

print(mean_Z, mean_ZH, mean_ZA)
print(Z.var(),ZH.var(),ZA.var())
```

```python
np.random.seed(12345)
N = 10000
Z = np.random.normal(size=(2,N))
cov_z = np.cov(Z[0,:],Z[1,:])  ### np.cov()
corr_z = np.corrcoef(Z[0,:],Z[1,:])  ### np.corrcoef()
cov_Z = np.cov(Z)
corr_Z = np.corrcoef(Z)
plt.plot(Z[0,:],Z[1,:],'.')
plt.axis('equal')
plt.show()
print(cov_z, corr_z, cov_Z, corr_Z)
```

```python
C = [[1.,-0.7],[-0.7,1.]]
L = np.linalg.cholesky(C) # Cholesky decomposition
N = 10000
Z = Z
Z1 = L @ Z
cov_z1 = np.cov(Z1)
corr_z1 = np.corrcoef(Z1[0,:],Z1[1,:])
corr_Z1 = np.corrcoef(Z1)
plt.plot(Z1[0,:],Z1[1,:],'.')
plt.axis('equal')
plt.show()
print(cov_z1,corr_z1,corr_Z1)
```
